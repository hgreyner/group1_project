```python
# Plot Heatmap
gmaps.configure(api_key=key)
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=10)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Country</dt><dd>{Countries}</dd>
<dt>Temp</dt><dd>{Average Temperature}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
info = [info_box_template.format(**row) for index, row in finilized_temp.iterrows()]
locations = finilized_temp[["Latitude", "Longtitude"]]
```

```python
# Add marker layer ontop of heat map
fig = gmaps.figure()

markers = gmaps.marker_layer(locations, info_box_content=info)

fig.add_layer(markers)

fig

# Display Map
```
